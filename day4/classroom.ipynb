```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
x1 = [2 , 2 ,10 , 10 , 2]
y1 = [2 , 6 , 6 , 2 , 2]
x2 = [2 , 3 , 10, 10 , 2]
y2 = [6, 8, 8, 6 , 6]
x3 = [3 , 4]
y3 = [8 , 6]
x4 = [4 , 4]
y4 = [6 , 2]
x5 = [6 , 6 , 8 , 8, 6]
y5 = [3.5 , 4.5, 4.5 , 3.5, 3.5]
circle = Circle([3, 7] , 0.3)
plt.gca().add_patch(circle)
plt.plot(x1, y1)
plt.plot(x2, y2, 'r')
plt.plot(x3, y3, 'b')
plt.plot(x4, y4)
plt.plot(x5, y5)
plt.savefig('my_graph.png', dpi=300, bbox_inches='tight')
plt.show()
```
