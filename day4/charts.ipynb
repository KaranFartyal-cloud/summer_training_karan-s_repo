```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

x2 = [3,6,9, 12, 15]
y2 = [4, 7, 10, 13, 16]

plt.figure(figsize=(8, 6))
plt.plot(x,y , marker = 'o' , linestyle = '-' , label = "normal")
plt.plot(x2, y2, marker='x' , color = 'y', linestyle = "--", label="3 3 ke gap main")
plt.legend()

for xi, yi in zip(x,y):
    plt.annotate(f'({xi} , {yi})', (xi, yi), ha='center')



plt.show()
```

```python
x = np.linspace(0 , np.pi * 3, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label = "sin x")
plt.plot(x, y_cos, label = "cost x")
plt.xlabel("X values")
plt.ylabel("Function value")
plt.title("Trignometric functions")
plt.legend()

plt.show()
```
