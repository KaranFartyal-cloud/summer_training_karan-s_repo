```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("woman.jpg")
img = np.array(img)

result = img.copy()

h, w, _ = result.shape

mid_h = h // 2
mid_w = w // 2

# # Top-left (Red)
result[:mid_h, :mid_w, 1] = 0   # Green
result[:mid_h, :mid_w, 2] = 0   # Blue

# Top-right (Blue)
result[:mid_h, mid_w:, 0] = 0   # Red
result[:mid_h, mid_w:, 1] = 0   # Green

# Bottom-left (Green)
result[mid_h:, :mid_w, 0] = 0   # Red
result[mid_h:, :mid_w, 2] = 0   # Blue


  
plt.imshow(result)
plt.axis("off")
plt.show()
```
